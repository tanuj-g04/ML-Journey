```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.regularizers import L2
```

```python
df_train=pd.read_csv("train.csv")
dftest=pd.read_csv("test.csv")
```

```python
display(df_train.head())
```

```python
columns=['Survived', 'Pclass', 'Sex', 'Age','SibSp','Parch','Fare']
dfmodel=df_train[columns].copy()
display(dfmodel.head(10))
```

```python
dfmodel['sexb']=dfmodel['Sex'].map({'male':1,'female':0})
dfmodel=dfmodel.drop('Sex',axis=1)
display(dfmodel.head(10))
```

```python
dfmodel['Age']=dfmodel['Age'].fillna(dfmodel['Age'].median())
dfmodel['Age'].isna().sum()
display(dfmodel.head(10))
dfmodel.corr()['Survived']
```

```python
y=df_train['Survived']
x=dfmodel.drop('Survived',axis=1)
scaler=StandardScaler()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
display(x_train)
```

```python
model=Sequential([
    Dense(units=40,activation='relu',kernel_regularizer=L2(0.01)),
    Dense(units=20,activation='relu',kernel_regularizer=L2(0.01)),
    Dense(units=30,activation='relu',kernel_regularizer=L2(0.01)),
    Dense(units=10,activation='relu',kernel_regularizer=L2(0.01)),
    Dense(units=1,activation='sigmoid')
])
model.compile(loss=tf.keras.losses.BinaryCrossentropy(),optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),metrics=["accuracy"])
model.fit(x_train,y_train,epochs=50,validation_data=(x_test,y_test))
```

```python
model.evaluate(x_test,y_test)
```

```python
dftest=dftest.sort_values(by='PassengerId')
dftest['Age']=dftest['Age'].fillna(dftest['Age'].median())
dftest['sexb']=dftest['Sex'].map({'male':1,'female':0})
dftest.isna().sum()
dftest['Fare']=dftest['Fare'].fillna(dftest['Fare'].median())
columns=['Pclass', 'Age','SibSp','Parch','Fare', 'sexb']
xpred=dftest[columns]
xpred_scaled=scaler.transform(xpred)
xpred.info()
display(dftest.head())
```

```python
predictions=model.predict(xpred_scaled)
binpred=(predictions>0.5).astype(int)
finalsub=pd.DataFrame({
    'PassengerId':dftest['PassengerId'],
    'Survived':binpred.flatten()
})
print(finalsub.head())
finalsub.to_csv("Titatnicv2.csv",index=False)
```
